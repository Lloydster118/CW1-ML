## CNN Breast_cancer dataset

```python
import numpy as np
import time
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import f1_score, classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, Dropout

start = time.time()

a = 0
for i in range(1000):
    a += (i**100)

# Load the Breast Cancer Wisconsin dataset
data = load_breast_cancer()
X = data.data  # Features
y = data.target  # Target labels (0 = Malignant, 1 = Benign)

# Reshape data to fit CNN input (samples, height, width, channels)
X = X.reshape(-1, 30, 1, 1)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Standardize the dataset
scaler = StandardScaler()
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)
X_train_flat = scaler.fit_transform(X_train_flat)
X_test_flat = scaler.transform(X_test_flat)
X_train = X_train_flat.reshape(X_train.shape)
X_test = X_test_flat.reshape(X_test.shape)

# Build a simple CNN model
model = Sequential([
    Conv2D(16, kernel_size=(1, 1), activation='relu', input_shape=(30, 1, 1)),
    Flatten(),
    Dense(64, activation='relu'),  # First hidden layer
    Dropout(0.3),
    Dense(32, activation='relu'),  # Second hidden layer
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, validation_split=0.2, epochs=50, batch_size=16, verbose=1)

# Evaluate the model
y_pred = (model.predict(X_test) > 0.5).astype(int).flatten()
f1_cnn = f1_score(y_test, y_pred)

print(f"CNN F1 Score (Breast Cancer): {f1_cnn:.2f}")

# Classification Report
print("\nClassification Report for CNN:")
print(classification_report(y_test, y_pred, target_names=['Malignant', 'Benign']))

end = time.time()
print("The time of execution of above program is :",
      (end-start) * 10**3, "ms")
```
